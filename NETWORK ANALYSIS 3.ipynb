## FACEBOOK DATASET

Importing the required libraries. 

```python
import pandas as pd
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

Loading the facebook dataset using pandas library

```python
df1=pd.read_csv(r"D:\360Digitmg\ASSIGNMENTS\Ass11\facebook.csv")
df1.head()
```

Converting the dataframe into a  numpy matrix

```python
df1_matrix=df1.to_numpy()
df1_matrix
```

converting the matrix into a graph object

```python
from networkx import from_numpy_matrix, Graph
g1=from_numpy_matrix(df1_matrix, create_using=Graph())
```

Creating  a circular network for facebook. 

```python
pos = nx.circular_layout(g1)
plt.figure(figsize=(25,12))
nx.draw_networkx(g1, pos, node_size = 200,cmap=plt.cm.Blues,with_labels=False)
```
